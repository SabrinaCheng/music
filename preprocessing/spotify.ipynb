```python
def get_songs(query, size):
    album_name = []
    album_uri = []
    album_release_date = []
    artist_name = []
    artist_uri = []
    track_name = []
    track_uri = []
    track_popularity = []
    audio_tempo = []

    for i in range(0, size, 50):
        _track = sp.search(q=query, limit=50, offset=i)['tracks']['items']
        for t in _track:
            # album name
            album_name.append(t['album']['name'])
            # album uri
            album_uri.append(t['album']['uri'])
            # album release date
            album_release_date.append(t['album']['release_date'])
            # artist name
            artist_name.append(t['album']['artists'][0]['name'])
            # artist uri
            artist_uri.append(t['album']['artists'][0]['uri'])
            # track name
            track_name.append(t['name'])
            # track uri
            track_uri.append(t['uri'])
            _track_uri = t['uri']
            # track popularity
            track_popularity.append(t['popularity'])

            # audio tempo
            audio_tempo.append(sp.audio_features(_track_uri)[0]['tempo'])
            
    
    track_dataframe = pd.DataFrame({
                                'album_name': album_name, 
#                                 'album_uri': album_uri,
                                'album_release_date': album_release_date,
                                'artist_name': artist_name,
#                                 'artist_uri': artist_uri,
                                'track_name': track_name,
#                                 'track_uri': track_uri,
                                'track_popularity': track_popularity,
                                'audio_tempo':audio_tempo
                               })
    print(track_dataframe.shape)
    display(track_dataframe)
```
